```python
from matplotlib import patches, pyplot as plt
import numpy as np
import seaborn as sns

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a subplot with 1 row and 2 columns, sharing the x-axis
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True)

# Number of lines
num_lines = 9

# Cyclic color palette
colors = sns.color_palette('husl', n_colors=num_lines)

# Plot on the first axis
for i in range(num_lines):
    axes[0].plot(x, y1 * (i + 1), label=f'Line {i + 1}', color=colors[i])

axes[0].set_title('Axis 1')

# Plot on the second axis
for i in range(num_lines):
    axes[1].plot(x, y2 * (i + 1), label=f'Line {i + 1}', color=colors[i])

axes[1].set_title('Axis 2')

# Label the shared x-axis
fig.text(0.5, 0.04, 'X-axis Label', ha='center', va='center')

# Get the last point of line one on ax1
last_point_ax1 = (x[-1], y1[-1])

# Get the first point of line two on ax2
first_point_ax2 = (x[0], y2[0])

arrow = patches.ConnectionPatch(
    last_point_ax1,
    first_point_ax2,
    coordsA=axes[0].transData,
    coordsB=axes[1].transData,
    # Default shrink parameter is 0 so can be omitted
    color=colors[0],
    arrowstyle="-",  # "normal" arrow
    mutation_scale=10,  # controls arrow head size
    linewidth=1,
    linestyle=':'
)

fig.patches.append(arrow)

plt.show()
```
